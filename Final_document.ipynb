The figure shown above (*Difference in GHG Emissions 2019/2020*) visualizes for each country the difference in their greenhouse gas emissions between the pandemic year 2020 and the pre-pandemic year 2019. \
\
It can be observed in the figure, that defining the variation in greenhouse gas emissions for each European country between the pre-pandemic and pandemic year, can be attributed to the concept of relative change. Countries with higher absolute emission values naturally have a greater margin for reduction, and thus, their emissions may register a more noticeable decline. This is due to the fact that a proportionally smaller reduction in absolute emissions for these countries can still translate into a substantial decrease in terms of actual volume. In contrast, countries with initially lower absolute emission values might experience a significant percentage decrease but contribute less to the overall reduction in greenhouse gas emissions. Therefore, the magnitude of decline is influenced by the baseline emissions of each country.\
\
The results also show that Spain has a more noticable decline in emissions in comparison to the other so-called 'big polluters'. It can be seen that Spain has a greater decrease due to the Covid-19 pandemic than France, Italy and Poland. This is quite noteworthy because these three countries have higher absolute greenhouse gas emissions.
